```python

try:
  sayi = int(input("Bir sayı gir=>"))
  print(sayi)
except ValueError:
  print("ALERT=> Girilen veri sayı değildir!!")
else:
  print("BASARILI=> Giriş işlemi başarılı")
finally:
  print("FINISH=> Program sonlandırıldı.")
```

```python

try:
  s = 7 / 0
  degisken = int(s)
except ZeroDivisionError:
  print("DONUSTURME=> Dönüşüm Hatası")
else:
  print("ALERT=> işlem tamamlandı", degisken)
finally:
  print("FINISH=> program bitti")
```

```python
while True:
  girdi = input("Bir sayı gir(Çıkmak için 'q' bas)")
  if girdi.lower() == '5':
    print("Programdan Çıkılıyor")
    break
  try:
    sayi = float(girdi)
  except ValueError:
    print("Hata: Girilen veri sayı değildir!!")
    continue

  if sayi == 0:
    print("Sayı sıfırdır")
  elif sayi>0:
    print("Sayı pozitif")
  else:
    print("Sayı negatif")

  break
```
